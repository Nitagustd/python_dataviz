```python
from matplotlib import pyplot as pp
import matplotlib
import pandas as pd
import numpy as np
```

```python
iris_PATH = 'C:\\Users\\User\\Downloads\\iris.data'
iris_data = pd.read_csv(iris_PATH ,header = None)
data_np = iris_data.iloc[:,:-1].to_numpy()
```

```python
onlyseto = iris_data.iloc[:50,:-1].to_numpy()
onlyversi = iris_data.iloc[50:100,:-1].to_numpy()
onlyvirgi = iris_data.iloc[100:,:-1].to_numpy()
```

```python
matplotlib.rcParams['figure.figsize'] = [15,15]
```

#pp.subplot(1,4,3)
#1 = แถวของsubplot
#4 = หลักของsubplot
#3 = ตำแหน่งของgraph

```python
pp.subplot(1,3,1)
setobox = pp.boxplot(onlyseto)
pp.subplot(1,3,2)
versibox = pp.boxplot(onlyversi)
pp.subplot(1,3,3)
virgibox = pp.boxplot(onlyvirgi)
```

##### ไม่ควรทำแบบนี้ เพราะมันมีประโยชน์แค่เทียบสเกล feature เฉย

```python
pp.subplot(1,4,1)
setobox1 = pp.boxplot(onlyseto[:,0])
pp.subplot(1,4,2)
setobox2 = pp.boxplot(onlyseto[:,1])
pp.subplot(1,4,3)
setobox3 = pp.boxplot(onlyseto[:,2])
pp.subplot(1,4,4)
setobox4 = pp.boxplot(onlyseto[:,3])
```

### ให้ดูง่ายๆ

```python
feat1 = (onlyseto[:,0],onlyversi[:,0],onlyvirgi[:,0])
feat2 = (onlyseto[:,1],onlyversi[:,1],onlyvirgi[:,1])
feat3 = (onlyseto[:,2],onlyversi[:,2],onlyvirgi[:,2])
feat4 = (onlyseto[:,3],onlyversi[:,3],onlyvirgi[:,3])
```

```python
matplotlib.rcParams['figure.figsize'] = [15,10]
```

```python
pp.boxplot(feat1,labels = ['seto','versi','virgi'],vert=False)
pp.title('feature1');
```

```python
pp.subplot(2,2,1)
pp.boxplot(feat1,labels = ['seto','versi','virgi'])
pp.title('feature1')
pp.subplot(2,2,2)
pp.boxplot(feat2,labels = ['seto','versi','virgi'])
pp.title('feature2')
pp.subplot(2,2,3)
pp.boxplot(feat3,labels = ['seto','versi','virgi'])
pp.title('feature3')
pp.subplot(2,2,4)
pp.boxplot(feat4,labels = ['seto','versi','virgi'])
pp.title('feature4')
```

## ดูการกระจายของข้อมูล(plot ตามพื้นที่ที่ตั้งของ data)

```python
import pandas as pd
```

```python
debt_df = pd.read_excel('Average_Debt.xls',skiprows = 3)
```

```python
debt_df
```

```python
dbd =debt_df[['Region and province',2549]]
```

```python
dbd.drop([0,1,2,3,8,31,49,70,85,86,87,88],axis = 0, inplace = True)
```

```python
dbd.head()
```

```python
import folium
```

```python
d2plot = dbd
```

```python
country_geo = 'thai.json'
```

```python
n_col = list()
for i in d2plot['Region and province']:
    n_col.append(i.lstrip())
d2plot['Region and province']= n_col
```

```python
map = folium.Map(location = [10,100],zoom_star = 5)
```

```python
map
```

```python
map.choropleth(geo_data = country_geo, data = d2plot,columns = ['Region and province',2549],
                key_on = 'feature.properties.name',fill_color = 'YlGnBu', fill_opacity = 0.5, legend_name = 'Average Dept')
```

```python
map.save('plot_data.html')
```
